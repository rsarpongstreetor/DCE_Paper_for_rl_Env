## Returns

```python
''''The retuns are calculated, States are available, the two forms the independent variable and the action space are also presented the actioin space ( dependent variable) is the price direction
action space is the direction of the returns and the returns is the the differnce of the fuel price form aprevious chprious price. The reward  is defind as the absolut of the returns
x=
'''
x=np.array(y)
def returns(x):
  import numpy as np
  import array as arr
  import pandas as pd
  a =np.mean(x,axis=0)
  yx=[]
  xx=[]
  xx= np.vstack((a,x))
  xx=np.array(xx)
  K=xx.shape[0]
  L=xx.shape[1]
  yxxxx=np.zeros((K-1,L))
  for i in range(L):
   for k in range(K-1):
    yxxxx[k-1][i]=(xx[k+1,i]-xx[k,i])

  print(yxxxx)
  return yxxxx

RRRR=returns(x)
print(RRRR)
RRRR.shape
```

PDF  And CDF of Fuel Price Returns

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
gh=RRRR.shape[1]
for i in range(gh):
    data=RRRR[:,i]
    count, bins_count = np.histogram(data, bins=10)

    pdf = count / sum(count)

    # using numpy np.cumsum to calculate the CDF
    # We can also find using the PDF values by looping and adding
    cdf = np.cumsum(pdf)

    # plotting PDF and CDF
    plt.grid()
    plt.title("Returns distribution", fontsize=10, y=1)
    plt.xlabel("Return")
    plt.ylabel("Probability")
    plt.plot(bins_count[1:], pdf, color="red", label="PDF")
    plt.plot(bins_count[1:], cdf, label="CDF")
    plt.legend()
    plt.show()
```

## Reward computation

```python
''''The retuns are calculated, States are available, the two forms the independent variable and the action space are also presented the actioin space ( dependent variable) is the price direction
action space is the direction of the returns and the returns is the the differnce of the fuel price form aprevious chprious price. The reward  is defind as the absolut of the returns


x=
'''
x=np.array(y)
def reward(x):
  import numpy as np
  import array as arr
  import pandas as pd
  a =np.mean(x,axis=0)
  yx=[]
  xx=[]
  xx= np.vstack((a,x))
  xx=np.array(xx)
  K=xx.shape[0]
  L=xx.shape[1]
  yxx=np.zeros((K-1,L))
  for i in range(L):
   for k in range(K-1):
    yxx[k-1][i]=abs(xx[k+1,i]-xx[k,i])

  print(yxx)
  return yxx

RRw=reward(x)
print(RRw)
RRw.shape
```

## Calculating Action Space

```python
#There are three action space which are up(+1), no change(0) and down(-1).
#We convert them to a (0,1, 2) space using the equation below  by addding 1.
#Apply actionss
        # 0 -1 = -1
        # 1 -1 = 0
        # 2 -1 = 1
"áctionssss finds the sign of the element of the matrix number"
def actionspace(x):
  import numpy as np
  import array as arr
  import pandas as pd
  a =np.mean(x,axis=0)
  yx=[]
  xx=[]
  xx= np.vstack((a,x))
  xx=np.array(xx)
  K=xx.shape[0]
  L=xx.shape[1]
  yxx=np.zeros((K-1,L))
  for i in range(L):
   for k in range(K-1):
    yx=xx[k+1,i]-xx[k,i]
    if yx>0:
      yxx[k-1][i]=2
    elif yx<0:
      yxx[k-1][i]=0
    else:
      yxx[k-1][i]=1
  print(yxx)
  return yxx

action= actionspace(RRw)
print(action)
action.shape


```

## Calculating Probability of Action space









```python
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import time
import sklearn
import collections
from collections import Counter
K=action.shape[0]
L=action.shape[1]
ps=[]
def probabilityaction(action):
  for i in range(L):
    ps.append(Counter(action[:,i]))
  pss=pd.DataFrame(ps)
  psss=pss.div(K)
  #print(psss)
  probaction=np.zeros((K,L))
  for i in range(len(ps)):
     for j in range(K):
        if action[j,i] == 0:
          probaction[j,i] = psss.iat[i,2]
        elif action[j,i] ==1:
          probaction[j,i] =psss.iat[i,0]
        else:
          probaction[j,i]=psss.iat[i,1]
        j += 1
     i+=1
  return probaction
```
